```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np

data = pd.read_csv('./summary.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")

epochs = [i for i in range(len(data.index))]

fig, ax = plt.subplots( dpi = 200, figsize=(10, 4))
plt.title('Loss / Epoch')
ax.plot(epochs, data['train_loss'], label="S1-Training Loss")
ax.plot(epochs, data['eval_loss'], label="S1-Validation Loss")
ax.set_xlabel('Epochs')
ax.set_ylabel('Loss')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots( dpi = 200, figsize=(10, 4))
plt.title('Top-k Error / Epoch')
ax.plot(epochs, data['eval_top1'], label="S1 Top1 Error")
ax.plot(epochs, data['eval_top5'], label="S1 Top5 Error")
ax.set_xlabel('Epochs')
ax.set_ylabel('Error')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots( dpi = 200, figsize=(10, 4))
plt.title('LR / Epoch')
ax.plot(epochs, data['lr'], label="S1 LR")
ax.set_xlabel('Epochs')
ax.set_ylabel('LR')
ax.legend()
plt.show()
```
